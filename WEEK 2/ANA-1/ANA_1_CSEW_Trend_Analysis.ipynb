multi-line MH trend chart

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(14,8))

years = [
    "2019-2020",
    "2021-2022",
    "2022-2023",
    "2023-2024"
]

# Plot each indicator
for i in range(len(selected_s43c)):

    values = selected_s43c.iloc[i,1:].values

    plt.plot(
        years,
        values,
        marker='o',
        linewidth=2,
        label=selected_s43c.iloc[i,0]
    )

# Title and labels
plt.title(
    "Mental Health Related Trend Analysis (2019–2024)",
    fontsize=18
)

plt.xlabel("Year", fontsize=13)

plt.ylabel("Prevalence Rate (%)", fontsize=13)

plt.xticks(rotation=0)

plt.legend(
    bbox_to_anchor=(1.05,1),
    loc='upper left'
)

plt.grid(True)

plt.tight_layout()

plt.show()
```

covid related spike graph

```python
covid_data = selected_s43c.iloc[0]

years = [
    "2019-2020",
    "2021-2022",
    "2022-2023",
    "2023-2024"
]

values = covid_data[1:].values

plt.figure(figsize=(8,5))

plt.plot(
    years,
    values,
    marker='o',
    linewidth=3
)

# Highlight COVID period
plt.axvspan(
    0.8,
    1.2,
    alpha=0.3,
    color='red',
    label='COVID Impact Period'
)

plt.title(
    "COVID-era Impact on Domestic Abuse Trend",
    fontsize=15
)

plt.xlabel("Year")

plt.ylabel("Prevalence Rate (%)")

plt.legend()

plt.grid(True)

plt.show()
```

bar chart


```python
import matplotlib.pyplot as plt
import numpy as np

years = [
    "2019-2020",
    "2021-2022",
    "2022-2023",
    "2023-2024"
]

domestic_abuse = [6.06, 5.73, 5.14, 5.35]
partner_abuse = [4.47, 3.92, 3.44, 3.55]
family_abuse = [2.13, 2.44, 2.15, 2.12]

x = np.arange(len(years))
width = 0.25

plt.figure(figsize=(12,6))

plt.bar(
    x - width,
    domestic_abuse,
    width,
    label='Domestic Abuse'
)

plt.bar(
    x,
    partner_abuse,
    width,
    label='Partner Abuse'
)

plt.bar(
    x + width,
    family_abuse,
    width,
    label='Family Abuse'
)

plt.xticks(x, years)

plt.ylabel("Prevalence Rate (%)")

plt.xlabel("Year")

plt.title(
    "Mental Health Related Indicator Trends (2019–2024)"
)

plt.legend()

plt.grid(axis='y')

plt.show()
```
